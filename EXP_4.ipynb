```python
'''Problem Statement 1:Unique Event Participants.
A college event has a list of students who registered multiple times by mistake.
Write a program using set to store participant names and display only unique participants'''


participants=["Ajay","Bob","Charan","Derik","Ajay","Eren","Ajay","Bob","Farhan","Ajay"]
unique_participants=set(participants)
print("Unique Participants:",unique_participants)
#list1=set9=frozenset(list1)  also this can be used and then list1 can be printed as a set
```

```python
'''Problem Statement 2:Common Subjects
Three students study multiple subjects.
Write a program to find the common subject'''

set_student1={"maths","python","physics","biology"}
set_student2={"python","physics","BEEE","english"}
set_student3={"sanskrit","hindi","python","english"}
print("Common Subjects:",set_student1 & set_student2 & set_student3)
```

```python
'''Problem Statement 3: One set contains students in cricket club and another in football club.
Find students who are :1. In both clubs 2. In only one club'''

set_cricket={"Siddhant","Shrirang","Shreyas"}
set_football={"Siddhant","Soham","Samarth"}
print("In both clubs:",set_cricket&set_football)
print("In only cricket club:",set_cricket-set_football)
print("In only football club:",set_football-set_cricket)
print("Only one club:",set_cricket^set_football)
```

```python
'''Problem Statement 4: Absent Students
Give a set of aa students and a set of present students,find the absent students.'''

setstd={"Sid","Bob","Sam","Ram","Roy"}
presentstd={"Sid","Sam"}
print("Absent Students:",setstd-presentstd)
```

```python
'''Problem Statement 5:Remove Invalid Entries.
A set contains valid course codes.
Remove a discontinued course code from the set.'''

subject={"Maths100","Ecsf101","Python102"}
subject.discard("Ecsf101")                       #We can also use subject.remove()
print(subject)
```
